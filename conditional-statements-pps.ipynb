# **looping statements are used to repeat a block of code multiple times**

# For loop
## For loop is used to iterate over a sequence (like a list, string, tuple, or range) and execute a block of code multiple times.

for variable in sequence:

    # code block

## 1. Using range()

```python
for i in range(5):
    print(i)
```

## 2. Loop Through a List

```python
studentnames = ["suresh ", "mahesh ", "rana"]

for i in studentnames:
    print(i)
```

## 3.3: Loop Through a String

```python
for letter in "Python":
    print(letter)
```

## 4: Using range(start, stop, step)

```python
for i in range(1, 10, 1):
    print(i)
```

```python
for i in range(1, 10, 3):
    print(i)
```

# Advance looping using for 

```python
for i in range(1, 6): #start from 1 to 5
    print("*" * i) # prints multiple stars
```

```python
for i in range(5, 0, -1): # start from 5 to 1 
    print("*" * i)
```

# While loop
The while loop in Python is used to repeat a block of code as long as a condition is True.

while condition:

    # code block

```python
i = 1
while i < 5: #condition
    print(i)
    i += 1
```

#  Loop Control Statements

## 1. break

Stops the loop completely.

```python
for i in range(5):
    if i == 3: # break execution here
        break
    print(i)
```

## 2. continue

Skips current iteration and moves to next.

```python
for i in range(5):
    if i == 2:
        continue #skip current iteration
    print(i)
```

## 3. pass

Does nothing (placeholder).

```python
for i in range(5):
    pass
```

```python
# with functions
def my_function():
    pass
```
